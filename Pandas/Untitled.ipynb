## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

unique_item = len(df['Item Name'].unique())
avg_price = '${:,.2f}'.format(df['Price'].mean())
purchases = df['Item Name'].count()
revenue = '${:,.2f}'.format(df['Price'].sum())

data = {'Number of Unique Items': unique_item,'Average Price': avg_price, 'Total Number of Purchases': purchases, 'Total Revenue': revenue}
df = pd.DataFrame([data], columns= data.keys())
df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

# sorting by SN 
df.sort_values("SN", inplace = True) 
  
# dropping ALL duplicte values 
df.drop_duplicates(subset ="SN", inplace = True) 

male_count = len([x for x in df["Gender"] if x == "Male"])
female_count = len([x for x in df["Gender"] if x == "Female"])
other_count = len([x for x in df["Gender"] if x == "Other / Non-Disclosed"])
total_count = len(df['SN'].unique())


d = {'Total Count': [male_count, female_count, other_count],'Percentage of Players': ['{:,.2f}'.format(int(male_count)*100/int(total_count)), '{:,.2f}'.format(int(female_count)*100/int(total_count)), '{:,.2f}'.format(int(other_count)*100/int(total_count))] }
df = pd.DataFrame(data=d, index = ['Male','Female','Other / Non-Disclosed'])
df
```

```python
#Another way
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

total_count = df.groupby("Gender")["SN"].nunique()
df = pd.DataFrame({"Total Count": total_count})
df["Percentage of Players"] = (total_count*100/total_count.sum()).map('{:,.2f}'.format)
df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

total_count = df.groupby("Gender")['Purchase ID'].nunique()
df1 = pd.DataFrame({"Purchase Count": total_count})
df1["Average Purchase Value"] = ((df.groupby("Gender")['Price'].mean())).map('${:,.2f}'.format)
df1["Total Purchase Value"] = ((df.groupby("Gender")['Price'].sum())).map('${:,.2f}'.format)
df1["Avg Total Purchase per Person"] = ((df.groupby("Gender")['Price'].sum())/df.groupby("Gender")["SN"].nunique()).map('${:,.2f}'.format)
df1
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

bins = [0,10,15,20,25,30,35,40,100]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-40","40+"]


df["Age Range"] = pd.cut(df["Age"], bins, right = False, labels=group_names)
age_range = df.groupby("Age Range")["Purchase ID"].count()
df1 = pd.DataFrame({"Purchase Count": age_range})
df1["Average Purchase Price"] = ((df.groupby("Age Range")['Price'].mean())).map('${:,.2f}'.format)
df1["Total Purchase Value"] = ((df.groupby("Age Range")['Price'].sum())).map('${:,.2f}'.format)
df1["Avg Total Purchase per Person"] = ((df.groupby("Age Range")['Price'].sum())/df.groupby("Age Range")["SN"].nunique()).map('${:,.2f}'.format)
df1
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

total_count = df.groupby("SN")["Item ID"].nunique()
df1 = pd.DataFrame({"Purchase Count": total_count})
df1["Average Purchase Price"] = ((df.groupby("SN")['Price'].mean())).map('${:,.2f}'.format)
df1["Total Purchase Value"] = ((df.groupby("SN")['Price'].sum())).map('${:,.2f}'.format)

df1.sort_values("Purchase Count", axis = 0, ascending = False, 
                 inplace = True, na_position ='last') 
df1.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
file = "Resources/purchase_data.csv"
df = pd.read_csv(file)

total_count = df.groupby(["Item ID", "Item Name"])
df1 = pd.DataFrame({"Purchase Count": total_count["Item ID"].count()})
df1["Item Price"] = (total_count['Price'].mean()).map('${:,.2f}'.format)
df1["Total Purchase Value"] = (total_count['Price'].sum()).map('${:,.2f}'.format)

df1.sort_values("Purchase Count", axis = 0, ascending = False, 
                 inplace = True, na_position ='last')
df1.head()
```

## Most Profitable Items
